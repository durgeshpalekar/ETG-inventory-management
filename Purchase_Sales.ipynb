```python
import time 
time.ctime()


id = int(input("Enter Product ID : ")) 
quant = int(input("Enter the quantity : "))
st = str(id)
for i in st:
    if st in rec and quant < rec[st]['quantity']:
        print("WELCOME",)
        
        print("Product ID :",(st))
        print("Product :",(rec[st]["name"]))
        print("Category: ",(rec[st]["category"]))
        print("Price: ",(rec[st]["price"]))
        print("Quantity: ",(quant))
        print("Expiry Date: ",(rec[st]["expdate"])) 
        print("------------------------")
        print("Billing Amout: ", quant * (rec[st]["price"]))
        (rec[st]["Quantity"]) = (rec[st]["quantity"]) - quant
        print(time.ctime())
        print("*")
        break


    else: 
        print("Sorry, the product ID or quantity does not exist. Enter a valid product ID or quantity .")
        break



    
```
